## Filter the data for movies released in the 1990s

```python
movies_1990s = netflix_df[(netflix_df['type'] == 'Movie') &
                        (netflix_df['release_year'] >= 1990) &
                        (netflix_df['release_year'] < 2000)]
```

```python
movies_1990s.head()
```

## Convert duration to numeric

```python
movies_1990s['duration'] = pd.to_numeric(movies_1990s['duration'], errors = 'coerce')
movies_1990s.describe()
```

### errors = 'coerce' ensures the numberic values in the duration column are replaced NaN during conversion

# Find the most frequent movie duration

```python
duration = int(movies_1990s['duration'].mode().iloc[0])  # Save result as 'duration'
print(f"The most frequent movie duration in the 1990s: {duration} minutes")
```

### The mode() method to find the most common values()

# Count the number of short action movies

```python
short_action_movies = movies_1990s[(movies_1990s['duration'] < 90) &
                                  (movies_1990s['genre'].str.contains('Action', na = False))]
short_movie_count = short_action_movies.shape[0]

print(f"The number of short action movies released in the 1990s: {short_movie_count}")
```

### shape[0] count the number of entries (no. of row in dataframe)
